```python
def check_null_empty(df,column,typeconversion=''):
    df[column]=df[column].astype(str)
    df[column]=df[column].apply(lambda x: x.replace('*','',x.count('*')))
    df=df[~(df[column].isnull())]
    df=df[~(df[column]=='')]
    if typeconversion!='':
        print(typeconversion)
        df[column]=df[column].astype(typeconversion)
    return df
```
