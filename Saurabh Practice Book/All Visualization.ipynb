```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x=np.linspace(5,100,100)
y=np.linspace(10,1000,100)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("X-Y Graph")
plt.xlim(20,80)
plt.ylim(200,800)
plt.show()
```

```python
plt.plot([1,4,6,8],[3,8,3,7],'r*')
plt.show()
```

```python
x=np.linspace(0,5,10)
y=np.linspace(3,6,10)
plt.plot(x,y,'g-',x,y**2,'r*',x,y**3,'b^')
plt.show()
```

```python
### SUBPLOTS
x=np.linspace(1,10,100)
y=np.log(x)

plt.figure()
#subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.title("y=log(x)")
plt.plot(x,y)

plt.subplot(122)
plt.title('y=log(x)**2')
plt.plot(x,y**2)
plt.show()
```

```python
plt.figure(1)
plt.subplot(221)
plt.title('linear')
plt.plot(x,x)

plt.subplot(2,2,2)
plt.title('Cubic')
plt.plot(x,x**3)

plt.figure(2)
plt.subplot(221)
plt.title("Exponential")
plt.plot(x,x**2)
plt.xlabel('Hello')
plt.ylabel('Why')

plt.subplot(222)
plt.title("log")
plt.plot(x,np.log(x))

plt.show()
```

```python
loan_df=pd.read_csv('loan.csv')
len(loan_df['id'].unique())==len(loan_df)
loan_df.info
```

```python
plt.figure(1)
plt.subplot(221)
plt.title("Funded Amount Investment")
plt.boxplot(loan_df['funded_amnt_inv'])
plt.subplot(222)
plt.title("Funded Amount")
plt.boxplot(loan_df['funded_amnt'])
plt.figure(2)
plt.subplot(221)
plt.title('Installement')
plt.boxplot(loan_df['installment']) 
plt.subplot(222)
plt.title("Loan Amount")
plt.boxplot(loan_df['loan_amnt'])
plt.show()
```

```python
plt.subplot(221)
plt.title("Funded Amount Investment")
plt.hist(loan_df['funded_amnt_inv'])
plt.subplot(222)
plt.title("Funded Amount")
plt.hist(loan_df['funded_amnt'])
plt.subplot(223)
plt.title('Installement')
plt.hist(loan_df['installment']) 
plt.subplot(224)
plt.yscale('log')
plt.title("Loan Amount")
plt.hist(loan_df['loan_amnt'])
plt.show()
```

```python
### SEABORN Module

sns.set_style('whitegrid')
sns.distplot(loan_df['funded_amnt_inv'])
plt.show()
### ************** UNIVARIATE VARIABLE ***************** ###
```

```python
sns.distplot(loan_df['funded_amnt_inv'][:200],rug=True)
plt.show()
```

```python
plt.subplot(221)
plt.title('Funded Amount Invested')
sns.distplot(loan_df['funded_amnt_inv'][:200],rug=True)
plt.subplot(222)
plt.title("Funded Amount")
sns.distplot(loan_df['funded_amnt'],hist=False)
plt.subplot(223)
plt.title('Installement')
sns.distplot(loan_df['installment']) 
plt.subplot(224)
plt.title("Loan Amount")
sns.boxplot(y=loan_df['loan_amnt'])
plt.show()
```

```python
sns.jointplot('funded_amnt','installment', loan_df,kind='reg') 
plt.show()
```

```python
sns.jointplot('funded_amnt','funded_amnt_inv', loan_df[(loan_df.funded_amnt<30000)&(loan_df.funded_amnt_inv<30000)],kind='hex',color='r') 
plt.show()
```
