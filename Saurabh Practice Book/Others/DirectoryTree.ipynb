```python
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.plot(np.cos(2*np.pi*np.arange(0.0,5.0,.02)),'r--')
plt.title('COS GRAPH')
plt.subplot(122)
plt.plot(np.sin(2*np.pi*np.arange(0.0,5.0,.02)),'g--')
plt.title('SIN GRAPH')
```
