```python
import pandas as pd
import numpy as np
import sklearn
```

```python
doc_train=pd.read_csv('movie_review_train.csv')
doc_train.head()
```

```python
doc_test=pd.read_csv('movie_review_test.csv')
doc_test.head()
```

```python
numpy_array = doc_train.as_matrix()
#print(numpy_array)
X = numpy_array[:,1]
Y = numpy_array[:,0]
Y = Y.astype('int')
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
vec = CountVectorizer(stop_words='english')
vec.fit(X)
print(type(vec))
print(len(vec.vocabulary_))
print(vec.get_feature_names())
print(len(vec.get_feature_names()))
```

```python
vec = CountVectorizer(stop_words='english',min_df=.03,max_df=.8)
vec.fit(X)
print((vec.get_feature_names()))
```
